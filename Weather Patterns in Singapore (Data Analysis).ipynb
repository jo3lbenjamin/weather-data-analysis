### Singapore's Climate (Factors Affecting Singapore's Weather)
Geographically, Singapore is located 1 degree north of the equator. Because of this, we enjoy a tropical climate where there are only 2 seasons, dry season and wet season. Since Singapore is very close to the equator, sunlight is very intense and it is usually hot and humid. 

Before we move on to the data analysis, I would like to firstly address the monsoon seasons that Singapore experiences. The Northeast Monsoon occurs from December to early March, and the Southwest Monsoon from June to September. In the Northeast Monsoon, Singapore experience 2 phases, the wet phase where it happens from December to January (The wet phase witnesses continuous moderate to heavy rainfall in the afternoons and early evenings) and a dry phase where it happens from February to March (The dry phase has little or no rain, it is windy). 

The second monsoon that Singapore experiences is called the Southwest monsoon. In this period, there are frequent showers and thunderstorm activity between predawn to midday but thunderstorms only last only for about half an hour during this period. Sumatra Squalls are also common in this period. According to weather.gov.sg, a sumatra squall refers to an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of Singapore. These Sumatra squalls usually hit Singapore in the early mornings for an hour or two, followed by light showers or cloudy conditions until the afternoon. This season experiences spells of dry weather.

Between the Northeast monsoon and the Southwest monsoon, Singapore experiences the inter-monsoon period. This happens from March-May and October-November.  In this period, there are showers in the afternoon and early evening. In the months of May and July, it is usually more hot and dry whereas in October and November, rainy weather is more frequent. 

With this information, we are now able to view the data and see how these factors affect Singapore's climate.

Reference:https://www.guidemesingapore.com/business-guides/immigration/get-to-know-singapore/weather-and-climate-in-singapore

````python
re1982 = "^1982"
df_all_1982 = df_merged[df_merged['month'].str.contains(re1982)] 

re2020 = "^2020"
df_all_2020 = df_merged[df_merged['month'].str.contains(re2020)].reset_index(drop=True)

nmonths=['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun','Jul', 'Aug', 'Sep','Oct', 'Nov', 'Dec']
````

````python
#print(df_all_2020[['mean_temp','temp_mean_daily_max','temp_mean_daily_min','max_temperature','temp_extremes_min']])
#print(df_all_1982[['max_temperature','temp_extremes_min']])

def lineGraphTemperature(dataFrame, title, year):  
    sns.set_style("darkgrid")
    ax = dataFrame.plot(x="month", y="mean_temp", label="Mean Temperature",figsize=(20,10))
    dataFrame.plot(x="month", y="temp_mean_daily_max", label="Mean Daily Maximum Temperature", ax=ax)
    dataFrame.plot(x="month", y="temp_mean_daily_min", label="Mean Daily Minimum Temperature", ax=ax)
    dataFrame.plot(x="month", y="max_temperature", label="Absolute Extreme Maximum Temperature", ax=ax)
    dataFrame.plot(x="month", y="temp_extremes_min", label="Absolute Extreme Minimum Temperature", ax=ax)
    
    if (year==2020):
        ymax = dataFrame["max_temperature"].max()
        xmax = dataFrame["max_temperature"].idxmax()
        ax.annotate(f'Maximum Temperature {dataFrame["max_temperature"].max()}', xy=(xmax, ymax), xytext=(xmax-1, ymax+1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
        ymin = dataFrame["temp_extremes_min"].min()
        xmin = dataFrame["temp_extremes_min"].idxmin()
        ax.annotate(f'Minimum Temperature {dataFrame["temp_extremes_min"].min()}', xy=(xmin, ymin), xytext=(xmin+1, ymin-1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            )
    elif (year==1982):
        ymax = dataFrame["max_temperature"].max()
        xmax = dataFrame["max_temperature"].idxmax()
        ax.annotate(f'Maximum Temperature {dataFrame["max_temperature"].max()}', xy=(xmax, ymax), xytext=(xmax-1, ymax+1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
        ax.annotate(f'Maximum Temperature {dataFrame["max_temperature"].max()}', xy=(xmax+1, ymax), xytext=(xmax+1, ymax+1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
        ax.annotate(f'Maximum Temperature {dataFrame["max_temperature"].max()}', xy=(xmax+9, ymax), xytext=(xmax+8, ymax+1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
        ymin = dataFrame["temp_extremes_min"].min()
        xmin = dataFrame["temp_extremes_min"].idxmin()
        ax.annotate(f'Minimum Temperature {dataFrame["temp_extremes_min"].min()}', xy=(xmin, ymin), xytext=(xmin+1, ymin-1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            )
        ax.annotate(f'Minimum Temperature {dataFrame["temp_extremes_min"].min()}', xy=(xmin+7, ymin), xytext=(xmin+8, ymin-1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            )
        
    ax.set_ylim(20,36)
    ax.legend(loc='lower center', bbox_to_anchor=(0,1.02,1,0.2), ncol=3, fancybox=True, shadow=True)

    plt.title(title, fontsize=15, fontweight='bold',pad=70)
    plt.xlabel('Month')
    plt.ylabel('Temperatures in Â°C')
    plt.xticks(range(len(dataFrame.index)), nmonths)
    plt.show()

    
lineGraphTemperature(df_all_2020, "Temperature Range in 2020",2020)
lineGraphTemperature(df_all_1982, "Temperature Range in 1982",1982)
````
